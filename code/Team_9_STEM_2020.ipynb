```
# Importing Libraries
%matplotlib notebook
%matplotlib inline
import numpy as np
import scipy as sp
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sns.set()

#Imprting Our Data
data = pd.read_csv('http://webpage.pace.edu/fparisi/STEM-2020/Python/bop_data_retrivedbyian.csv')
```

```
dataa= pd.DataFrame(data)
```

```
# correlation Function
def corr(x,y):
  temp = data[[x, y]]
  dataT = temp[[x,y]].dropna()
  x1 = dataT[x]
  x2 = dataT[y]
  rho, pstat = stats.pearsonr(x1, x2)
  print('rho = %.4f' % rho, 'p-value = %.4f' % pstat)

```

```
# correlation oyster size and turbidity
corr("sizeOfLiveOysterMM","turbidity_average")
```

```
# correlation oyster size and salinity
corr("sizeOfLiveOysterMM","salinity_average")
```

```
# correlation oyster size and pH
corr("sizeOfLiveOysterMM","pH_average")
```
