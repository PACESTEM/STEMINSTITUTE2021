```
#(f - 32) * 5.0/9

filt = oyster_data['waterTemperature_units'] == "f"

def convert_temp(temp):
  return (temp - 32) * 5.0/9

oyster_data.loc[filt, "waterTemperature_average"] = oyster_data.loc[filt, "waterTemperature_average"].apply(convert_temp)

# lambda function makes code more concise, but harder to read. equivalent to convert_temp()

# df2["waterTemperature_average"] = df2["waterTemperature_average"].apply(lambda x: (x - 32) * 5.0/9)
#oyster_data.loc[filt]
del oyster_data["waterTemperature_units"]
oyster_data = oyster_data.dropna()
oyster_data
```

```
oyster_data.groupby("waterTemperature_average").mean()
```

```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

oyster_data.groupby("waterTemperature_average")["sizeOfLiveOysterMM"].mean().plot(kind="line", linewidth="2.5", marker="o")

plt.style.use("fivethirtyeight")
#oyster_data.plot(kind="scatter", x="waterTemperature_average", y="sizeOfLiveOysterMM")
plt.xlabel("Water Temperature (C deg)")
plt.ylabel("Average Size of Oyster (in mm)")
plt.savefig("plot1.png")
plt.show()
```

```
print(oyster_data.groupby("salinity_average").mean())

oyster_data.groupby("salinity_average")["sizeOfLiveOysterMM"].mean().plot(kind="line", marker="o", linewidth="2.5")

plt.style.use("fivethirtyeight")
#oyster_data.plot(kind="scatter", x="waterTemperature_average", y="sizeOfLiveOysterMM")
plt.ylabel("Mean Size of Oyster (mm)")
plt.savefig("plot2.png")
plt.show()
```
